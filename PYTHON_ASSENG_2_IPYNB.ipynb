2. Explain the difference between mutable and immutable data types with examples

Ans:-**Mutable data type -** Mutable data types are objects whose value can be changed in place after creation. This means you can add, remove, or change elements without creating a new object in memory.
  example-   
          my_list = [1, 2, 3]
              my_list.append(4)
              my_list is now [1, 2, 3, 4]
 **Immutable data type-**
            Immutable data types are objects whose value cannot be changed after they are created.
      exampe-
            my_string = "hello"
            new_string = my_string + " world"
            my_string is still "hello"
            a new object, new_string, holds "hello world"

5. Why might you use a set instead of a list in Python ?
Ans:-You might use a set instead of a list in Python for a few key reasons, primarily when you need to store a collection of unique items and require fast look-up times. Sets are also useful when the order of elements doesn't matter.

viewed_products_list = ['milk', 'bread', 'milk', 'cheese', 'bread', 'cheese']

    if product not in unique_products_list:
