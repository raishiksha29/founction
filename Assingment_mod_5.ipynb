Q.20 What is the difference between file.read() and file.readline() in Python ?

Explanation -
           n Python, both file.read() and file.readline() are methods used to read data from a file, but they differ in how much data they read at a time.

      #  file.read() -
                     The file.read() method reads the entire content of a file into a single string. You can also pass an optional argument to specify the number of characters or bytes to read.

      # file.readline() -
                        The file.readline() method reads a single line from a file, up to and including the newline character (\n).

     
      # Example -

                   # Create a sample file
             with open("sample.txt", "w") as f:
                  f.write("Line 1\n")
                  f.write("Line 2\n")

                  # Using read()
            with open("sample.txt", "r") as f:
                  content = f.read()
                  print("Content from read():")
                  print(content) # Output: Line 1\nLine 2\n

            # Using readline()
              with open("sample.txt", "r") as f:
                  line1 = f.readline()
                  print("\nContent from readline():")
                  print(line1) # Output: Line 1\n
                  line2 = f.readline()
                  print(line2) # Output: Line 2\n

Explanation -
            Memory management in Python is handled automatically by a private heap and garbage collection, but it still presents several challenges for developers. These challenges are often related to understanding how Python manages memory under the hood and how that can impact performance, especially in long-running or memory-intensive applications

      # challenges -
                 
                 * Memory Leak

                 * High Memory Overhead

                 * Fragmentation

                 * Lack of Explicit Control
    # Example -

                   import gc

             class Node:
                 def __init__(self, name):
                 self.name = name
                 self.next = None

             # Create two objects that reference each other
               a = Node("Node A")
               b = Node("Node B")
               a.next = b
               b.next = a

             # The external references are removed, but the objects remain in a cycle
             del a
             del b

            # The objects are now unreachable but not garbage-collected because of the cycle
            # The garbage collector's cycle detector will eventually collect them,
            # but in complex scenarios, this may not always happen immediately or at all.
            gc.collect()
