Q.1 What is Object-Oriented Programming (OOP) ?

Explanation -
                **OOPS** is a programming paradigm that organizes software design around objects rather than functions and logic.
                An object is a self-contained unit that bundles data (attributes) and the procedures that operate on that data (methods). 
                This approach aims to model real-world entities, making code more modular, reusable, and easier to manage.
    
Example-  

    # class (blueprint)
      class bike:
         def_inti_(self ,brand , model):
            self. brand = brand  #attribute
            self.model = model


          def display_info(self):  # Method 
             return f"bike: {self.brand} {self.model}"

         # obeject ( instance of class )
           bike1 = bike ("TCS" , "MODEL D")
           print(bike1.display_info())  # output bike:TCS MODEL D

Q.3 What is an object in OOP ?

Explanation : 
             In Object-Oriented Programming (OOP), an object is a specific instance of a class. It's a fundamental building block of an OOP system and represents a real-world or conceptual entity. Unlike a class, which is a blueprint, an object is a concrete, tangible entity that exists in a program's memory.
